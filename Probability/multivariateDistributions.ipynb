```python
N = 10000
x1 = np.random.normal(5, 4, N)
x2 = np.random.lognormal(.22, 1, N)
X = np.array([x1, x2]).T
```

```python
X_df = pd.DataFrame(X, columns=['X1', 'X2'])
X_columns = X_df.columns
```

```python
X_df.columns
```

```python
#def distributions(X_df):
fig, ax = plt.subplots()
for rv in X_columns:
    ax.hist(X_df[rv],alpha=.5, bins=20)
    
```
